```python
import pandas as pd
import pyarrow.parquet as pq

import datetime
```

```python
# import parquet files
train_click = pd.read_parquet('/content/drive/MyDrive/0.capstone/train_click.parquet')
train_cart = pd.read_parquet('/content/drive/MyDrive/0.capstone/train_cart.parquet')
train_order = pd.read_parquet('/content/drive/MyDrive/0.capstone/train_order.parquet')
```

```python
# function to filter by week. used query instead of loc as it is more memory efficient
def filter_by_week(df, start_unixtime, end_unixtime):
    return df.query('@start_unixtime <= ts < @end_unixtime').reset_index(drop=True)
```

```python
train_click_1 = filter_by_week(train_click, 1659304800, 1659909600)
train_click_1.to_parquet('/content/drive/MyDrive/0.capstone/train_click_1.parquet')
```
